```python
def convert_to_float(value):
    return float(value.replace("M", "").replace(",", "."))

df["Vol."] = df["Vol."].apply(convert_to_float)

print(df)
```

```python
def convert_to_floats(value):
    return float(value.replace("%", "").replace(",", "."))

df["Var%"] = df["Var%"].apply(convert_to_floats)

print(df)
```
